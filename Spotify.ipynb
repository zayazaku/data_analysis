Data frame transformation

```python
df['Total'] = df['liveness'] + df['valence']
df.head()
```

```python
df.drop(['playlist_url','track_id', 'artist_id', 'speechiness',
         'acousticness', 'instrumentalness', 'liveness', 
        'valence', 'tempo', 'Total'], 
        axis = 1, inplace = True)
df.head()
```

data visualization

```python
import matplotlib.pyplot as plt
duration_by_genre = df.groupby('year')['duration_ms'].mean()
duration_by_genre.plot(kind = 'bar')
plt.show()
```

```python
import matplotlib.pyplot as plt
duration_by_genre = df.groupby('year')['duration_ms'].mean()
duration_by_genre.plot(kind = 'line')
plt.show()
```

```python
#scatter
df.plot(x = 'danceability', y = 'loudness', kind = 'scatter', title = 'Danceability VS Loudness')
plt.show()
```

```python
#histogram 1
df[df['artist_name'] == 'Coldplay']['track_popularity'].hist()
#histogram 2
df[df['artist_name'] == 'Linkin Park']['track_popularity'].hist()
plt.legend(['Coldplay', 'Linkin Park'])
plt.show()
```
