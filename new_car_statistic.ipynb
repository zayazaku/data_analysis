Pearson Correlation

```python
data.corr()
```

```python
#correlation heatmap of dataset
def correlation_heatmap(df):
    _ , ax = plt.subplots(figsize =(14, 12))
    colormap = sns.diverging_palette(220, 10, as_cmap = True)
    
    _ = sns.heatmap(
        df.corr(), 
        cmap = colormap,
        square=True, 
        cbar_kws={'shrink':.9 }, 
        ax=ax,
        annot=False, 
        linewidths=0.1,vmax=1.0, linecolor='white',
        annot_kws={'fontsize':12 }
    )
    
    plt.title('Pearson Correlation of Features', y=1.05, size=15)
correlation_heatmap(data)
```

```python
from scipy import stats
```

```python
data = data.dropna()
```

HorsePower Vs Price

```python
pearson_coef, p_value = stats.pearsonr(data['Horsepower'], data['Price_in_thousands'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)  
```

Curb-weight VS Price

```python
pearson_coef, p_value = stats.pearsonr(data['Curb_weight'], data['Price_in_thousands'])
print('The pearson Correlation Coefficient is', pearson_coef, 'with a P-value of P =', p_value)
```

Engine size VS Price

```python
pearson_coef, p_value = stats.pearsonr(data['Engine_size'], data['Price_in_thousands'])
print('The pearson Correlation Coefficient is', pearson_coef, 'with a P-value of P =', p_value)
```
